```python
import json
```

```python
data = json.load(open('/content/CHallenge2-datascience-etl/challenge2-data-science-LATAM-main/TelecomX_Data.json'))
```

```python
# prompt: Para comenzar, realiza un anÃ¡lisis descriptivo de los datos, calculando mÃ©tricas como media, mediana, desviaciÃ³n estÃ¡ndar y otras medidas que ayuden a comprender mejor la distribuciÃ³n y el comportamiento de los clientes.
# ðŸ“Œ Consejos:
# ðŸ”— DocumentaciÃ³n de DataFrame.describe()

import pandas as pd

# Convert the data to a pandas DataFrame
df = pd.DataFrame(data)

# Perform descriptive analysis
descriptive_stats = df.describe()

# Print the descriptive statistics
descriptive_stats
```

```python
# prompt: for df need to unnest the dictionaries in customer, phone, internet and account columns

customer_df = pd.json_normalize(df['customer'])
phone_df = pd.json_normalize(df['phone'])
internet_df = pd.json_normalize(df['internet'])
account_df = pd.json_normalize(df['account'])

# Concatenate the new dataframes with the original one and drop the original nested columns
df = pd.concat([df, customer_df, phone_df, internet_df, account_df], axis=1).drop(columns=['customer', 'phone', 'internet', 'account'])

df.head()
```

```python
# prompt: expand disctionary within a column is a dataframe, eval() arg 1 must be a string, bytes or code object, avoid that error

#import json
#import pandas as pd
# Assuming you have already cloned the repository and loaded the data as shown in the preceding code
# !git clone https://github.com/javierDDT/CHallenge2-datascience-etl.git
# data = json.load(open('/content/CHallenge2-datascience-etl/challenge2-data-science-LATAM-main/TelecomX_Data.json'))
# df = pd.DataFrame(data)

def safe_loads(x):
  try:
    return json.loads(x)
  except (json.JSONDecodeError, TypeError):
    # Return an empty dictionary or a dictionary with default values
    # depending on how you want to handle invalid data
    return {}

customer_expanded = df['customer'].apply(safe_loads).apply(pd.Series)
account_expanded = df['account'].apply(safe_loads).apply(pd.Series)
phone_expanded = df['phone'].apply(safe_loads).apply(pd.Series)
internet_expanded = df['internet'].apply(safe_loads).apply(pd.Series)
```

```python
# prompt: En este paso, el objetivo es comprender cÃ³mo estÃ¡ distribuida la variable "churn" (evasiÃ³n) entre los clientes. Utiliza grÃ¡ficos para visualizar la proporciÃ³n de clientes que permanecieron y los que se dieron de baja.

import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of the 'churn' variable
plt.figure(figsize=(6, 4))
sns.countplot(x='Churn', data=df)
plt.title('Distribution of Churn')
plt.xlabel('Churn')
plt.ylabel('Number of Customers')
plt.show()

# Visualize the proportion of churn
churn_counts = df['Churn'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'salmon'])
plt.title('Proportion of Churn')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# prompt: Ahora, exploraremos cÃ³mo se distribuye la evasiÃ³n segÃºn variables categÃ³ricas, como gÃ©nero, tipo de contrato, mÃ©todo de pago, entre otras.
# Este anÃ¡lisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudarÃ¡ a orientar acciones estratÃ©gicas.

import matplotlib.pyplot as plt
# Analyze churn distribution by categorical variables
categorical_cols = ['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
                    'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
                    'PaperlessBilling', 'PaymentMethod']

for col in categorical_cols:
    plt.figure(figsize=(8, 5))
    sns.countplot(x=col, hue='Churn', data=df)
    plt.title(f'Churn Distribution by {col}')
    plt.xlabel(col)
    plt.ylabel('Number of Customers')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# prompt: Ahora, exploraremos cÃ³mo se distribuye la evasiÃ³n segÃºn variables categÃ³ricas, como gÃ©nero, tipo de contrato, mÃ©todo de pago, entre otras.
# Este anÃ¡lisis puede revelar patrones interesantes, por ejemplo, si los clientes de ciertos perfiles tienen una mayor tendencia a cancelar el servicio, lo que ayudarÃ¡ a orientar acciones estratÃ©gicas.

import matplotlib.pyplot as plt
# Visualize churn distribution by gender
plt.figure(figsize=(8, 5))
sns.countplot(x='gender', hue='Churn', data=df)
plt.title('Churn Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')
plt.show()

# Visualize churn distribution by contract type
plt.figure(figsize=(8, 5))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn Distribution by Contract Type')
plt.xlabel('Contract Type')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualize churn distribution by payment method
plt.figure(figsize=(10, 6))
sns.countplot(x='PaymentMethod', hue='Churn', data=df)
plt.title('Churn Distribution by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Analyze the distribution of numerical variables for churned and non-churned customers

numerical_cols = ['tenure', 'MonthlyCharges', 'TotalCharges'] # Assuming these are the numerical columns

for col in numerical_cols:
  plt.figure(figsize=(10, 6))
  sns.histplot(data=df, x=col, hue='Churn', kde=True, common_norm=False)
  plt.title(f'Distribution of {col} by Churn')
  plt.xlabel(col)
  plt.ylabel('Frequency')
  plt.show()
```
